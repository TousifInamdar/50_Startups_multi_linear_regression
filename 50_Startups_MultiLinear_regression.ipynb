```python
#Store state
state=data["State"]
```

```python
#Will use LabelEncoder to convert strings into some labels like 0,1,2
from sklearn.preprocessing import LabelEncoder
```

```python
state_le=LabelEncoder()
```

```python
state_le
```

```python
#Fit and transforms all data of state into Labels
state_le_final=state_le.fit_transform(state)
```

```python
#Labels are as New_Work=2 , California=0 , Florida=1
state_le_final
```

```python
#Converts 1D array into 2D array
state_le_final_2d = state_le_final.reshape(-1,1)
```

```python
state_le_final_2d.shape
```

```python
state_le_final_2d
```

```python
#Create Dummy columns to make different columns for each state using OneHot Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
state_ohe=OneHotEncoder()
```

```python
state_ohe
```

```python
#Fit and transforms
state_dump = state_ohe.fit_transform(state_le_final_2d)
```

```python
state_dump
```

```python
#Converts Into Array
state_final=state_dump.toarray()
```

```python
state_final
```

```python
# dummy variable trap: Multicollinarity
#Eliminats 3rd Column
# every state: Independent Variable : into model
state_final_trap = state_final[: , 0:2]
```

```python
state_final_trap
```

```python
#Drops State column
data=data.drop(columns=['State'])
```

```python
data
```

```python
#Concatenate data and state_final_trap : first convert numpy array into pandas DataFrame:
type(state_final_trap)
```

```python
#Converts numpy array into pandas DataFrame
state_pandas=pandas.DataFrame(state_final_trap,columns=['California','Florida'])
```

```python
state_pandas
```

```python
#Concatenat horizontally
data=pandas.concat([data ,state_pandas], axis=1)
```

```python
data
```

```python
y=data["Profit"]
```

```python
#Modify X variable with states i.e. California and NewYork
X=data[['R&D Spend', 'Administration', 'Marketing Spend', 'California' , 'Florida']]
```

```python
#Fits The Model
model.fit(X,y)
```
